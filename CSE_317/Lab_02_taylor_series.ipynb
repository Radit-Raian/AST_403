## Taylor Series:

The Taylor series expansion of a function $f(x)$ about the point $x = a$ is given by:

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x - a)^n
$$
$$
f(x) = f(a) + \frac{f'(a)}{1!}(x - a) + \frac{f''(a)}{2!}(x - a)^2 + \frac{f^{(3)}(a)}{3!}(x - a)^3 + \cdots + \frac{f^{(n)}(a)}{n!}(x-a)^n
$$

where $f^{(n)}(a)$ is the $n$-th derivative of $f$ evaluated at $x = a$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=0;

def f(x):
    return np.exp(-x)
def f1(x):
    return -np.exp(-x)
def f2(x):
    return np.exp(-x)
def f3(x):
    return -np.exp(-x)
def f4(x):
    return np.exp(-x)

print (f(x))
print (f1(x))
print (f2(x))
print (f3(x))
print (f4(x))
```

```python
x=0;
n=100;

def fn(x,n):
    return (-1)**(n)*np.exp(-x)
print (fn(x,n))
```

```python
def factorial(n):
    if n==0:
       return 1
    return np.prod(np.arange(1,n+1,1))
print(factorial(5))
```

```python
def taylor(x,a,n):
    total=0.000

    for k in range (n+1):
        term=((-1)**k*np.exp(-a)*(x-a)**k/factorial(k))
        total+=term
    return total
```

```python
x=1;
taylor_approx = taylor(1,0,4)
actual_val = f(x)
absolute_diff=np.abs(taylor_approx-actual_val)
print(absolute_diff)
```

```python
x_val = np.arange(-2,2,0.1)
y_val = f(x_val)

orders=[1,2,3,4]
colors=['red','green','blue','black']
y_app=[]

for n in orders:
    y_approx=[taylor(x,0,n) for x in x_val]
    y_app.append(y_approx)

plt.plot(x_val, y_val, color='orange', linewidth=2)

for n, color in zip(orders, colors):
    y_approx = [taylor(x, 0, n) for x in x_val]
    plt.plot(x_val, y_approx, color=color, linestyle='dashed')

plt.title('Taylor Series Approximation')
plt.show()
```
